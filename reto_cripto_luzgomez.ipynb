```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
btc = yf.download("BTC-USD", start="2024-01-01", end="2024-12-31", interval="1d")
btc.head()
```

## 🐼 Inspección con pandas

```python
btc.head()
```

## 📈 Visualización con Plotly

```python
fig = px.line(btc, x=btc.index, y="Close", title="BTC-USD 2024")
fig.show()
```

## 🔥 Análisis multi-cripto

```python
# Descarga 4 criptos y compara su rendimiento %
tickers = ["BTC-USD", "ETH-USD", "SOL-USD", "ADA-USD"]
data = yf.download(tickers, start="2025-01-01")["Close"]
rend = (data.iloc[-1] / data.iloc[0] - 1) * 100
print(rend.sort_values(ascending=False))
```
